```python
import iolib
import os
```

```python
files = os.listdir("final_round_2016.in/")
```

```python
data = {f: iolib.read_input("final_round_2016.in/" + f) for f in files}
```

```python
data.keys()
```

```python
for f in files:
    figure()
    hist(data[f]["sats"][:, -2])
    title("%s w [%d sats]" % (f, len(data[f]["sats"])))
    savefig("w-%s.png" % f)
    figure()
    hist(data[f]["sats"][:, -1])
    title("%s d" % f)
    savefig("d-%s.png" %f)
    figure()
    hist([x[0] for x in data[f]["collections"]])
    title("%s V" % f)
    savefig("V-%s.png" %f)
    figure()
    hist([x[0] / len(x[1]) for x in data[f]["collections"]])
    title("%s V / L" % f)
    savefig("V-over-L-%s.png" %f)
    figure()
    hist([len(x[1]) for x in data[f]["collections"]])
    title("%s L" % f)
    savefig("L-%s.png" %f)
    
    
```
